# Inferencing
This cell demonstrated how we can use the model object as well as the created access token

## Defining the model id
We need to specify model id that will be used for inferencing:


```python
model_id = "meta-llama/llama-3-2-1b-instruct"
```

## Defining the model parameters
We need to provide a set of model parameters that will influence the
result:

```python
parameters = {
    "decoding_method": "greedy",
    "max_new_tokens": 200,
    "min_new_tokens": 0,
    "repetition_penalty": 1
}
```

## Defining the project id or space id
The API requires project id or space id that provides the context for the call. We will obtain
the id from the project or space in which this notebook runs:

```python
project_id = os.getenv("PROJECT_ID")
space_id = os.getenv("SPACE_ID")
```

## Defining the Model object
We need to define the Model object using the properties we defined so far:


```python
from ibm_watsonx_ai.foundation_models import ModelInference

model = ModelInference(
	model_id = model_id,
	params = parameters,
	credentials = credentials,
	project_id = project_id,
	space_id = space_id
	)
```

## Defining the inferencing input
Foundation model inferencing API accepts a natural language input that it will use
to provide the natural language response. The API is sensitive to formatting. Input
structure, presence of training steps (one-shot, two-shot learning etc.), as well
as phrasing all influence the final response and belongs to the emerging discipline of
Prompt Engineering.

Let us provide the input we got from the Prompt Lab:


```python
Type =  """text""" 
Summary =  """text""" 

prompt_input = f"""You are a YouTube content strategist and copywriter AI.

Your job is to take a long summary of a YouTube video and generate:
1. A catchy, high-performing YouTube **title** (max 60 characters)
2. A long, SEO-optimized **description** (5–10 lines)
3. Relevant and trending **hashtags** (8–12)

Guidelines:
- The **title** should be clear, emotionally engaging, curiosity-driven, and suitable for any content.
- The **description** must explain the video in detail, highlighting the main questions or themes covered.
- End the description with 8–12 **YouTube hashtags** on new lines (no commas).

Use this format exactly:

Title:  
<Generated title here>

Description:  
<Generated description here>

Hashtags:  
#hashtag1  
#hashtag2  
#hashtag3  
...

Input: Type: History  
Summary: A detailed account of the fall of the Roman Empire, including internal political instability, economic decline, invasions, and cultural shifts.

Output: Title:  
Why Did Rome Fall? The Collapse of an Empire

Description:  
What really caused the fall of the Roman Empire? Dive into the historical collapse driven by corruption, civil unrest, economic chaos, and barbarian invasions. This video unpacks the timeline, impact, and legacy of one of history’s greatest empires.

Hashtags:  
#RomanEmpire  
#FallOfRome  
#AncientHistory  
#HistoryExplained  
#WorldHistory  
#CivilizationCollapse  
#HistoricalEvents  
#RomeDecline  
#EmpireLegacy  
#PastToPresent


Input: Type: Finance  
Summary: A beginner-friendly guide to managing personal finances in your 20s. Topics include budgeting, saving, building credit, avoiding debt, and starting early investments to build long-term wealth.

Output: Title:
Master Your Money in Your 20s: Budget, Save & Grow

Description:
Are you in your 20s and unsure where to start with money? This beginner-friendly guide walks you through the financial foundations every young adult needs — from budgeting basics to building credit, saving smartly, and avoiding debt traps. Learn how small decisions today can lead to massive wealth tomorrow. Whether you're just starting your first job or looking to get financially organized, this video is your complete roadmap.

Hashtags:
#PersonalFinance
#MoneyTips
#BudgetingBasics
#FinancialFreedom
#CreditScore
#InvestSmart
#FinanceInYour20s
#MoneyMindset
#WealthBuilding
#MillennialFinance
#DebtFreeJourney
#SmartSaving


Type: Vlog  
Summary: A day in my life living in Tokyo, Japan. From navigating a rush-hour train, grabbing sushi in a hidden local spot, to ending the night at a peaceful shrine — this vlog shares a calm but exciting experience of city life.

Output"""
```

## Execution
Let us now use the defined Model object and pair it with input and
